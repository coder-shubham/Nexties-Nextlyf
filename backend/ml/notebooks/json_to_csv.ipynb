```python
data_path = "/Users/clarence/Documents/playground/HackBangalore/final-singapore/Nexties-Nextlyf/backend/ml/data/updated_data_with_names.json"

import json
import pandas as pd

with open(data_path, 'r') as f:
    data = json.load(f)
```

```python
room_data = []
event_data = []
for doc in data['response']['docs']:
    if 'Wifi' in doc.keys():
        room_data.append(doc)
    else:
        event_data.append(doc)
```

```python
room_df = pd.DataFrame(room_data)
room_df.to_csv('/Users/clarence/Documents/playground/HackBangalore/final-singapore/Nexties-Nextlyf/backend/ml/data/rooms.csv', index=False)
room_df.head()
```

```python
event_df = pd.DataFrame(event_data)
event_df.to_csv('/Users/clarence/Documents/playground/HackBangalore/final-singapore/Nexties-Nextlyf/backend/ml/data/events.csv', index=False)
event_df.head()
```
