## Rooms

```python
import pandas as pd

room_df = pd.read_csv('../data/updated_rooms.csv', usecols=['__id__', 'Type', 'Category', 'PropertyLocation',
       'Price', 'AmenitiesIncluded', 'Color', 'Wifi', 'AC', 'TV', 'Fridge', 'Bathroom', 'Balcony', 
       'Tags', 'No. of Beds', 'Bed Type', 'Floor No.', 'Rating/PopularityPercentage'])
room_df = room_df.replace('Yes', 1).replace('No', 0).replace('yes', 1).replace('no', 0)
room_df.head()
```

```python
room_df.columns
```

```python
# no. of beds
# print(room_df['No. of Beds'].value_counts())

room_df['No. of Beds'] = room_df['No. of Beds'].apply(lambda x: x / 3)
# print(room_df['No. of Beds'].value_counts())

# floor no.
# print(room_df['Floor No.'].value_counts())

room_df['Floor No.'] = room_df['Floor No.'].apply(lambda x: x / 25)
# print(room_df['Floor No.'].value_counts())

# popularity percentage
room_df['Rating/PopularityPercentage'] = room_df['Rating/PopularityPercentage'].apply(lambda x: x / 100)

# Category
room_df['Category'] = room_df['Category'].apply(lambda x: 4 if x == 'Luxury' else 3 if x == 'Premium' else 2 if x == 'Deluxe' else 1)
room_df['Category'] = room_df['Category'].apply(lambda x: x / 4)

# Type
room_df['Type'] = room_df['Type'].apply(lambda x: 1 if x == 'Shared Room' else 2 if x == 'Single Room' else 3 if x == 'Studio' else 4 if x == 'Loft' else 5)
room_df['Type'] = room_df['Type'].apply(lambda x: x / 5)

# Bed Type
room_df['Bed Type'] = room_df['Bed Type'].apply(lambda x: 1 if x == 'Single Bed' else 2 if x == 'Double Bed' else 3 if x == 'Queen Bed' else 4)
```

```python
extra_cols = []

# amenties
for element_list in room_df['AmenitiesIncluded'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

print(extra_cols)

# price
# room_df['Price'].describe()
price_breaks = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
for i in range(1, len(price_breaks)):
    room_df['Price_' + str(price_breaks[i-1]) + '_' + str(price_breaks[i])] = room_df['Price'].apply(lambda x: 1 if x > price_breaks[i-1] and x <= price_breaks[i] else 0)
    # extra_cols.append('Price_' + str(price_breaks[i-1]) + '_' + str(price_breaks[i]))
print(extra_cols)

# tags
for element_list in room_df['Tags'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)
print(extra_cols)

# property location
for element in room_df['PropertyLocation'].unique().tolist():
    extra_cols.append('PropertyLocation_' + element)
print(extra_cols)

# color
for element in room_df['Color'].unique().tolist():
    extra_cols.append('Color_' + element)

# # type
# for element in room_df['Type'].unique().tolist():
#     print(element)
#     extra_cols.append('Type_' + element)

# # bed type
# for element in room_df['Bed Type'].unique().tolist():
#     extra_cols.append('Bed Type_' + element)

# # category
# for element in room_df['Category'].unique().tolist():
#     extra_cols.append('Category_' + element)
```

```python
import pandas as pd

friends_df = pd.read_csv(
    '../data/user_details_new.csv', 
    usecols=['__id__', 'gender', 'dob', 'country', 'city', 
             'language', 'profession', 'fieldsOfInterest', 'hobbies']
)
friends_df.head()
```

```python
friends_df.columns
```

```python
extra_cols = []

# gender
for element in friends_df['gender'].unique().tolist():
    extra_cols.append('gender_' + element)

# country
for element in friends_df['country'].unique().tolist():
    extra_cols.append('country_' + element)

# city
for element in friends_df['city'].unique().tolist():
    extra_cols.append('city_' + element)

# language
for element in friends_df['language'].unique().tolist():
    extra_cols.append('language_' + element)

# profession
for element in friends_df['profession'].unique().tolist():
    extra_cols.append('profession_' + element)

# fields of interest
for element_list in friends_df['fieldsOfInterest'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

# hobbies
for element_list in friends_df['hobbies'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

# age group
# eg 19/04/95
friends_df['year'] = pd.to_datetime(friends_df['dob'], format='%d/%m/%y')
friends_df['age'] = 2024 - friends_df['year'].dt.year
age_breaks = [18, 25, 35, 45, 60]
for i in range(1, len(age_breaks)):
    friends_df['age_' + str(age_breaks[i-1]) + '_' + str(age_breaks[i])] = friends_df['age'].apply(lambda x: 1 if x > age_breaks[i-1] and x <= age_breaks[i] else 0)
    # extra_cols.append('age_' + str(age_breaks[i-1]) + '_' + str(age_breaks[i]))

print(extra_cols)
```
