## Rooms

```python
import pandas as pd

room_df = pd.read_csv('../data/rooms.csv', usecols=['id', 'Type', 'Category', 'PropertyLocation',
       'Price', 'AmenitiesIncluded', 'Color', 'Wifi', 'AC', 'TV', 'Fridge',
       'Bathroom', 'Balcony', 'Tags',
       'No._of_Beds', 'Bed_Type', 'Floor_No.', 'Rating_PopularityPercentage'])
room_df.head()
```

```python
room_df.columns
```

```python
extra_cols = []

# amenties
for element_list in room_df['AmenitiesIncluded'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

print(extra_cols)

# price
# room_df['Price'].describe()
price_breaks = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
for i in range(1, len(price_breaks)):
    room_df['Price_' + str(price_breaks[i-1]) + '_' + str(price_breaks[i])] = room_df['Price'].apply(lambda x: 1 if x > price_breaks[i-1] and x <= price_breaks[i] else 0)
    # extra_cols.append('Price_' + str(price_breaks[i-1]) + '_' + str(price_breaks[i]))
print(extra_cols)

# tags
for element_list in room_df['Tags'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)
print(extra_cols)

# property location
for element in room_df['PropertyLocation'].unique().tolist():
    extra_cols.append('PropertyLocation_' + element)
print(extra_cols)

# color
for element in room_df['Color'].unique().tolist():
    extra_cols.append('Color_' + element)

# type
for element in room_df['Type'].unique().tolist():
    extra_cols.append('Type_' + element)

# bed type
for element in room_df['Bed_Type'].unique().tolist():
    extra_cols.append('Bed_Type_' + element)

# category
for element in room_df['Category'].unique().tolist():
    extra_cols.append('Category_' + element)
```

```python
filtered_room_df_cols = ['id', 'Type', 'Category', 'Wifi', 'AC', 'TV', 'Fridge', 'Bathroom', 'Balcony', 
                         'No._of_Beds', 'Bed_Type', 'Floor_No.', 'Rating_PopularityPercentage',
                         'Price_0_50', 'Price_50_100', 'Price_100_150', 'Price_150_200', 'Price_200_250', 
                         'Price_250_300', 'Price_300_350', 'Price_350_400', 'Price_400_450', 'Price_450_500']
```

```python
len(filtered_room_df_cols + extra_cols)
```

```python
# create a new dataframe with filtered columns and extra columns
filtered_room_df = room_df[filtered_room_df_cols].copy()
for col in extra_cols:
    filtered_room_df[col] = 0

filtered_room_df.head()

# fill in the extra columns
for index, row in room_df.iterrows():
    for value in row['AmenitiesIncluded'].split(', '):
        filtered_room_df.at[index, value] = 1
    for value in row['Tags'].split(', '):
        filtered_room_df.at[index, value] = 1
    filtered_room_df.at[index, 'PropertyLocation_' + row['PropertyLocation']] = 1
    filtered_room_df.at[index, 'Color_' + row['Color']] = 1
    filtered_room_df.at[index, 'Type_' + row['Type']] = 1
    filtered_room_df.at[index, 'Bed_Type_' + row['Bed_Type']] = 1
    filtered_room_df.at[index, 'Category_' + row['Category']] = 1

# for index, row in room_df.iterrows():
#     for col in filtered_room_df.columns:
#         if col in extra_cols:
#             if col in row['AmenitiesIncluded'].split(', '):
#                 filtered_room_df.at[index, col] = 1
#             elif col in row['Tags'].split(', '):
#                 filtered_room_df.at[index, col] = 1
#             elif col in row['PropertyLocation']:
#                 filtered_room_df.at[index, 'PropertyLocation_'+col] = 1
#             elif col in row['Color']:
#                 filtered_room_df.at[index, 'Color_'+col] = 1
#             elif col in row['Type']:
#                 filtered_room_df.at[index, 'Type_'+col] = 1
#             elif col in row['Bed_Type']:
#                 filtered_room_df.at[index, 'Bed_Type_'+col] = 1
#             elif col in row['Category']:
#                 filtered_room_df.at[index, 'Category_'+col] = 1

filtered_room_df.head()
```

## Friends

```python
import pandas as pd

friends_df = pd.read_csv(
    '../data/user_details.csv', 
    usecols=['userId', 'gender', 'dob', 'country', 'city', 
             'language', 'profession', 'fieldsOfInterest', 'hobbies']
)
friends_df.head()
```

```python
friends_df.columns
```

```python
extra_cols = []

# gender
for element in friends_df['gender'].unique().tolist():
    extra_cols.append('gender_' + element)

# country
for element in friends_df['country'].unique().tolist():
    extra_cols.append('country_' + element)

# city
for element in friends_df['city'].unique().tolist():
    extra_cols.append('city_' + element)

# language
for element in friends_df['language'].unique().tolist():
    extra_cols.append('language_' + element)

# profession
for element in friends_df['profession'].unique().tolist():
    extra_cols.append('profession_' + element)

# fields of interest
for element_list in friends_df['fieldsOfInterest'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

# hobbies
for element_list in friends_df['hobbies'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

# age group
# eg 19/04/95
friends_df['year'] = pd.to_datetime(friends_df['dob'], format='%d/%m/%y')
friends_df['age'] = 2024 - friends_df['year'].dt.year
age_breaks = [18, 25, 35, 45, 60]
for i in range(1, len(age_breaks)):
    friends_df['age_' + str(age_breaks[i-1]) + '_' + str(age_breaks[i])] = friends_df['age'].apply(lambda x: 1 if x > age_breaks[i-1] and x <= age_breaks[i] else 0)
    # extra_cols.append('age_' + str(age_breaks[i-1]) + '_' + str(age_breaks[i]))

print(extra_cols)
```

```python
filtered_friends_df_cols = ['userId', 'age_18_25', 'age_25_35', 'age_35_45', 'age_45_60']
```

```python
len(filtered_friends_df_cols + extra_cols)
```

```python
# create a new dataframe with filtered columns and extra columns
filtered_friends_df = friends_df[filtered_friends_df_cols].copy()
# filtered_friends_df = pd.concat(filtered_friends_df, axis=1)
new_cols_df = pd.DataFrame(0, index=filtered_friends_df.index, columns=extra_cols)

# Concatenate the new columns to the existing DataFrame
filtered_friends_df = pd.concat([filtered_friends_df, new_cols_df], axis=1)
# filtered_friends_df = pd.concat([filtered_friends_df] + extra_cols, axis=1)
# for col in extra_cols:
#     filtered_friends_df.loc[:, col] = 0

# filtered_friends_df = pd.concat(filtered_friends_df, axis=1)

filtered_friends_df.head()

# fill in the extra columns
for index, row in friends_df.iterrows():
    for value in row['fieldsOfInterest'].split(', '):
        filtered_friends_df.at[index, value] = 1
    for value in row['hobbies'].split(', '):
        filtered_friends_df.at[index, value] = 1
    filtered_friends_df.at[index, 'gender_' + row['gender']] = 1
    filtered_friends_df.at[index, 'country_' + row['country']] = 1
    filtered_friends_df.at[index, 'city_' + row['city']] = 1
    filtered_friends_df.at[index, 'language_' + row['language']] = 1
    filtered_friends_df.at[index, 'profession_' + row['profession']] = 1

filtered_friends_df.head()
```
