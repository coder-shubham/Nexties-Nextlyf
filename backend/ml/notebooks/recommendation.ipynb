```python
import pandas as pd

room_df = pd.read_csv('../data/rooms.csv', usecols=['Type', 'Category', 'PropertyLocation',
       'Price', 'AmenitiesIncluded', 'Color', 'Wifi', 'AC', 'TV', 'Fridge',
       'Bathroom', 'Balcony', 'Tags',
       'No._of_Beds', 'Bed_Type', 'Floor_No.', 'Rating_PopularityPercentage'])
room_df.head()
```

```python
room_df.columns
```

```python
extra_cols = []

# amenties
for element_list in room_df['AmenitiesIncluded'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)

print(extra_cols)

# price
# room_df['Price'].describe()
price_breaks = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
for i in range(1, len(price_breaks)):
    room_df['Price_' + str(price_breaks[i-1]) + '_' + str(price_breaks[i])] = room_df['Price'].apply(lambda x: 1 if x > price_breaks[i-1] and x <= price_breaks[i] else 0)
    extra_cols.append('Price_' + str(price_breaks[i-1]) + '_' + str(price_breaks[i]))
print(extra_cols)

# tags
for element_list in room_df['Tags'].unique().tolist():
    for element in element_list.split(', '):
        if element not in extra_cols:
            extra_cols.append(element)
print(extra_cols)

# property location
for element in room_df['PropertyLocation'].unique().tolist():
    extra_cols.append('PropertyLocation_' + element)
print(extra_cols)

# color
for element in room_df['Color'].unique().tolist():
    extra_cols.append('Color_' + element)

# type
for element in room_df['Type'].unique().tolist():
    extra_cols.append('Type_' + element)

# bed type
for element in room_df['Bed_Type'].unique().tolist():
    extra_cols.append('Bed_Type_' + element)

# category
for element in room_df['Category'].unique().tolist():
    extra_cols.append('Category_' + element)
```

```python
filtered_room_df_cols = ['Type', 'Category', 'Wifi', 'AC', 'TV', 'Fridge', 'Bathroom', 'Balcony', 
                         'No._of_Beds', 'Bed_Type', 'Floor_No.', 'Rating_PopularityPercentage']
```

```python
len(filtered_room_df_cols + extra_cols)
```

```python
# create a new dataframe with filtered columns and extra columns
filtered_room_df = room_df[filtered_room_df_cols].copy()
for col in extra_cols:
    filtered_room_df[col] = 0

filtered_room_df.head()

# fill in the extra columns
for index, row in room_df.iterrows():
    for col in filtered_room_df.columns:
        if col in extra_cols:
            if col in row['AmenitiesIncluded']:
                filtered_room_df.at[index, col] = 1
            elif col in row['Tags']:
                filtered_room_df.at[index, col] = 1
            elif col in row['PropertyLocation']:
                filtered_room_df.at[index, col] = 1
            elif col in row['Color']:
                filtered_room_df.at[index, col] = 1
            elif col in row['Type']:
                filtered_room_df.at[index, col] = 1
            elif col in row['Bed_Type']:
                filtered_room_df.at[index, col] = 1
            elif col in row['Category']:
                filtered_room_df.at[index, col] = 1

filtered_room_df.head()
```
